```python
import os
import shutil


def createTestDatset(source_dir,destination_dir):
  # Define your source and destination directories
  # source_dir = '/content/Fire-Detection-1/train'
  # destination_dir = '/content/Fire-Detection-1/test'

  # Create the destination directory if it doesn't exist
  os.makedirs(destination_dir, exist_ok=True)

  # Get a list of all files in the source directory
  all_files = sorted(os.listdir(source_dir))

  # Get the last 2000 files
  files_to_move = all_files[-2000:]

  # Move the files
  for file_name in files_to_move:
      source_file = os.path.join(source_dir, file_name)
      destination_file = os.path.join(destination_dir, file_name)
      shutil.move(source_file, destination_file)

  print(f'Moved {len(files_to_move)} files to {destination_dir}')


createTestDatset('/content/Fire-Detection-1/train/images','/content/Fire-Detection-1/test/images')
```

```python
import cv2
import supervision as sv
from ultralytics import YOLO

model = YOLO(f'{HOME}/weights/yolov8l.pt')
image = cv2.imread(f'/content/data/sample_image1.jpg')
results = model(image)[0]
detections = sv.Detections.from_ultralytics(results)

bounding_box_annotator = sv.BoundingBoxAnnotator()
label_annotator = sv.LabelAnnotator()

annotated_image = bounding_box_annotator.annotate(
    scene=image, detections=detections)
annotated_image = label_annotator.annotate(
    scene=annotated_image, detections=detections)

sv.plot_image(annotated_image)
```
